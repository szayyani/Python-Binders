Commençons par l'importations des bibliothéques pertinente, ici $matplotlib$ pour les graphiques et $numpy$ pour le traitement des ensembles de données :

```python
import numpy as np
import matplotlib.pyplot as plt
```

Nous allons d'abord déclarer nos constantes et puis nos données

```python
  ####constantes
G = 6.67e-11 #cosntante universelle de gravitation
Msol = 1.98e30 #masse du soleil
Dsol = 1392539.0 #diametre du soleil

  ### Donnees
planets = ["mercure", "venus", "terre", "lune", "mars", "jupiter", "saturne", "uranus", "neptune"]
Diametre = np.array([4787, 12104, 12756, 3476, 6794, 142984, 120536,51118, 49922])*1e3 #diametre en m
distance = np.array([57.9, 108.2, 149.6, 149.6, 227.9, 778.3, 1427, 2871, 4497.1])*1e9 #distance en m
M = np.array([0.33, 4.87, 5.98, 0.0735, 0.642, 1899, 568, 86.8, 102])*1e24
Fsol = np.array([]) #création d'un array
gsol = np.array([])
```

Voici donc un exemple de code pour afficher la force gravitationnelle $F$ subie par chaque planete en fonction de sa distance par rapport au Soleil :

```python
for i in range(len(M)) :
    Fsol = np.append(Fsol, G*Msol*M[i]/(distance[i]**2))
    plt.plot(distance[i], Fsol[i], "o", label = planets[i])

for i in range(len(M)) :
    print("l'action du soleil sur ", planets[i]," est ", Fsol[i], " N")

plt.legend()
plt.xlabel("distance au soleil")
plt.ylabel("Force du soleil sur la planete")
plt.grid(True)
plt.title("Action du soleil sur les planetes en fonction de la distance")
plt.savefig( "planets.png")
plt.show()
```

On voit clairement que la force gravitationnelle du soleil varie selon la distance et la masse de la planete.


Just pour l'intérêt, on va faire un graphique aussi de l'intensité du champ gravitationnelle su soleil au niveau de chaque planète

```python
for i in range(len(M)) :
    gsol = np.append(gsol, Fsol[i]/M[i])
    plt.plot(distance[i], gsol[i], "o", label = planets[i])

plt.legend()
plt.xlabel("distance au soleil")
plt.ylabel("Force du soleil sur la planete")
plt.grid(True)
plt.title("Action du soleil sur les planetes en fonction de la distance")
plt.savefig( "planets.png")
plt.show()
```

Intéréssant de notre que le champ gravitationnelle diminue presque de manière exponentielle avec la distance.



Maintenant pour le calcul de l'intensité de pesenateur à la surface de chaque planète. On se rappelle la défintion mathématique de l'intensité de la pesanteur
$$ g = G\cdot ( \dfrac{M_{planète}}{R_{planète}^2}) = G\cdot ( \dfrac{4M_{planète}}{D_{planète}^2})$$

Et donc il suffit de faire une boucle pour faire ce calcul à partir des données


```python
gplanet = np.array([]) #création d'un array pour les valeur de pesanteur par planète

for i in range(len(M)) :
  gplanet = np.append(gplanet, G*M[i]*4/(Diametre[i]**2))
  print("l'intensité de la pesanteur à la surface de ", planets[i], " est ", gplanet[i], " N/kg")
```
