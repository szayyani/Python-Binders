```python
import math
```

Pour une equation du type : $ ax^2 + bx + c = 0 $

Entrez $a$ 



```python
a = float(input('Entre a '))
```

Entrez $b$

```python
b = float(input('Entre b '))
```

Entrez c 

```python
c = float(input('Entre c '))
```

Trouvons d'abord le discriminant 


```python
disc = b**2 - 4*a*c
print("le descriminant est = ", disc)
```

et les racines 

```python
root1 = (-b - math.sqrt(disc))/(2*a)
root2 = (-b + math.sqrt(disc))/(2*a)
```

et donc nous avons 

```python
print ("premier racine est : ", root1)
print ("deuxieme racine est : ", root2)
```
