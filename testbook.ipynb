Voici donc un premier test. On veut etudier les equations suivantes 
$$ z(t) = \dfrac{1}{2}gt^2 + v_xt$$


On commence par l'importation des bibliotheques qui nous interessent : 

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

et puis nous declarons nos constantes : 

```python
#### constants et conditions intiales
g = -9.81
```

et enfin les conditions initiales

```python
#### Conditions intiales
v0 = 10
theta_deg = 75 
theta_rad = theta_deg*np.pi/180
x0 = 0
y0 = 100

v0x = v0*np.cos(theta_rad)
v0y = v0*np.sin(theta_rad)
```

Definissions les liste des differentes variables 


```python
X=[]
Y=[]
Vx=[]
Vy=[]
T=[]
```


Definissons donc les fonctions qui sont nos equations horaires : 


```python
def x(t) : 
    return v0x*t + x0

def y(t) :
    return 0.5*g*t**2 + v0y*t + y0

def vx(t) : 
    return v0x

def vy(t) : 
  return g*t + v0y
```

et maintenant on va donner les points d'abcisses


```python
dt = 0.1    
#t_max=10   
#t = np.linspace(0,t_max, 20)
```

While loop until we touch the ground : 


```python
t = 0
y = y0
x = x0
while y>0 : 
  x = v0x*t + x0
  X.append(x)
  y = 0.5*g*t**2 + v0y*t + y0
  Y.append(y)
  vx = v0x
  Vx.append(vx)
  vy = g*t + v0y
  Vy.append(vy)
  T.append(t)
  t = t + dt
```

Visualisons le graphique 

```python
plt.plot(T,Y, label='y(t)')
plt.plot(T,X, label='x(t)')
plt.plot(T,Vx, label='Vx(t)')
plt.plot(T,Vy, label='Vy(t)')
plt.legend(ncol=2)
plt.show
print("late date ou la balle touche le sol est =",  t, "secondes")
```
