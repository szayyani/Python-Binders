Comment faire un graphique

```python
import numpy as np
import matplotlib.pyplot as plt

# définir nos séries de valeurs i et r

i = np.array([0,5,10,15,20,25,30,35,40,45,50,55,60,65,70,75,80]) #les angles d'incidence
r = np.array([0,4,8,12,16,20,23,26,29,32,35,38,40,42,44,46,48]) #les angles de refraction

# calculons les angles en radians

irad = np.radians(i)
rrad = np.radians(r)

# calculons le sinus de i et de r

sini = np.sin(irad)
sinr = np.sin(rrad)


#modélisation linéaire

a,b = np.polyfit(sinr, sini,1)

#visualisons le graphique

plt.scatter(sinr, sini, label="les sinus")
plt.plot(sinr, a*sinr + b, 'r')
plt.title("la réfraction")
plt.xlabel("r (degrée)")
plt.ylabel("i (degrée)")
plt.legend()
plt.grid('true')
plt.show()
```
