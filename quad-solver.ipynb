```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

On va definir une fonction qui prends comme arguments $a, b, c$ venant de la forme canonique de : $ ax^2 + bx + c = 0 $

```python
def quadsolver(a,b,c):
  disc = b**2 - 4*a*c
  sqrt_disc = math.sqrt(abs(disc))


  if disc > 0 :
    r1 = (-b - sqrt_disc)/2*a
    r2 = (-b + sqrt_disc)/2*a
    print("les racines sont reelles et distinctes")
    print("r1 = ", r1)
    print("r2 = ", r2)

  elif disc == 0 :
    r12 = -b/2*a
    print("les racines sont reelles et identiques")
    print("r12 = ", r12)

  elif disc < 0 :
    r12 = -b/2*a
    print("les racines sont complexes et distinctes")
    print("r1 = ", r12,"- i",sqrt_disc/(2*a) )
    print("r2 = ", r12,"+ i",sqrt_disc/(2*a))
```
